```python
graph = {
    'P': ['Q', 'R'],
    'Q': ['S'],
    'R': ['T', 'U'],
    'S': [],
    'T': ['V'],
    'U': [],
    'V': []
}

visited = set()

def dfs(node):
    if node not in visited:
        visited.add(node)
        print(node, end=" ")

        for neighbor in graph[node]:
            dfs(neighbor)

print("DFS Traversal:")
dfs('P')
```
